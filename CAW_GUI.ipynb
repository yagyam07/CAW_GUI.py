```python
%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
st.set_option('deprecation.showPyplotGlobalUse', False)

# Sample DataFrame
crime = pd.read_csv(r"/content/Crimes_Women_Final.csv")  # Assuming you have your DataFrame here


# Tab navigation
tabs = ["Home", "About Us", "Basic EDA", "Crimes Year wise","Crime Wise Proportion"]
selected_tab = st.sidebar.radio("Select Tab", tabs)

if selected_tab == "Home":
    st.title("Analysis of Crimes Against Women")
    st.image("/content/Crime_1.png", use_column_width=True)
    st.image("/content/Crime_2.jpg", use_column_width=True)
    st.image("/content/Crime_3.png", use_column_width=True)

elif selected_tab == "About Us":
    st.title("About Us")
    st.write("Welcome to our analysis of crimes against women. This application aims to provide insights into various crimes against women based on the provided dataset. Our goal is to raise awareness and contribute towards creating a safer environment for women.")
    st.write("Feel free to explore the different tabs to understand the data better and gain insights into trends and patterns.")
    st.image("/content/Crime_4.jpg", use_column_width=True)

elif selected_tab == "Basic EDA":
    st.title("Basic EDA")
    # Dropdown for selecting graph
    graph_choice = st.selectbox("Select Graph", ["Count of each Crime", "Count of Crimes Yearwise", "Trend over the years", "Heatmap", "Average number of cases by state"])

    if graph_choice == "Count of each Crime":
        crime_counts = crime[['Rape', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty',
                         'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']].sum()
        crime_counts_sorted = crime_counts.sort_values(ascending=False)

        plt.figure(figsize=(10, 6))
        ax = crime_counts_sorted.plot(kind='bar', color='skyblue', edgecolor='black')
        plt.xlabel('Crime Type')
        plt.ylabel('Total Occurrences')
        plt.title('Total Occurrences of Each Crime Type')
        plt.grid(True)

        # Positioning the x-axis labels below the bars
        ax.set_xticklabels(crime_counts_sorted.index, rotation=45, ha='right')

        for i, count in enumerate(crime_counts_sorted):
            ax.text(i, count + 1000, str(count), ha='center', va='bottom', fontsize=10)

        plt.tight_layout()
        st.pyplot()

    elif graph_choice == "Count of Crimes Yearwise":
        # Graph 2 code
        crime['sum'] = crime[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                      'Assault on women with intent to outrage her modesty',
                      'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
                      'Importation of Girls']].sum(axis=1)

        total_crime_by_year = crime.groupby('Year').sum()

        # Plot the line chart using Matplotlib directly
        plt.figure(figsize=(10, 6))
        plt.plot(total_crime_by_year.index, total_crime_by_year['sum'], marker='o', color='red', linewidth=2)
        plt.xlabel('Year')
        plt.ylabel('Total Crime Count')
        plt.title('Total Crime Count by Year')
        plt.grid(True)  # Add grid lines
        plt.tight_layout()

        # Display the plot using Streamlit
        st.pyplot()

    elif graph_choice == "Trend over the years":
        # Graph 4 code
        year_rape = crime.groupby('Year')['Rape'].sum().sort_values(ascending=False)
        year_kid = crime.groupby('Year')['Kidnapping and Abduction'].sum().sort_values(ascending=False)
        year_dowry = crime.groupby('Year')['Dowry Deaths'].sum().sort_values(ascending=False)
        year_cruelty = crime.groupby('Year')['Cruelty by Husband or his Relatives'].sum().sort_values(ascending=False)
        year_insult = crime.groupby('Year')['Insult to modesty of Women'].sum().sort_values(ascending=False)
        year_assault = crime.groupby('Year')['Assault on women with intent to outrage her modesty'].sum().sort_values(ascending=False)
        plt.figure(figsize=(5, 3))
        plt.plot(year_rape, label='Rape Cases')
        plt.plot(year_kid, label='Kidnapping Cases')
        plt.plot(year_dowry, label='Dowry Cases')
        plt.plot(year_cruelty, label='Cruelty Cases')
        plt.plot(year_insult, label='Insult Cases')
        plt.plot(year_assault, label='Assault Cases')
        plt.title("Year Wise Cases")
        plt.legend(loc='best')
        plt.xlabel("YEARS")
        plt.ylabel("NO OF CASES")
        plt.tight_layout()
        st.pyplot()

    elif graph_choice == "Heatmap":
        # Graph 5 code
        columns_of_interest = [
            'Rape',
            'Kidnapping and Abduction',
            'Dowry Deaths',
            'Assault on women with intent to outrage her modesty',
            'Insult to modesty of Women',
            'Cruelty by Husband or his Relatives',
            'Importation of Girls'
        ]
        crime_subset = crime[columns_of_interest]

        correlation_matrix = crime_subset.corr()

        plt.figure(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".4f", annot_kws={"size": 10})
        plt.title('Correlation Matrix for Crimes Against Women')
        st.pyplot()

    elif graph_choice == "Average number of cases by state":
        # Graph 6 code
        sunset_palette = sns.color_palette("deep", n_colors=10)
        average_rape_by_state = crime.groupby('STATE/UT')['Cruelty by Husband or his Relatives'].mean()
        plt.figure(figsize=(12, 6))
        average_rape_by_state.plot(kind='bar', color=sunset_palette)
        plt.xlabel('State')
        plt.ylabel('Average Number of Cruelty Cases')
        plt.title('Average Number of Cases by State')
        plt.xticks(rotation=90)
        plt.tight_layout()
        st.pyplot()

elif selected_tab == "Crimes Year wise":
    st.title("Crimes over the years")
    # Function to update the displayed text based on the selected state/UT
    def update_display():
        selected_state_ut = selected_state_ut_var
        total_sum_text = ""
        if selected_state_ut:
            total_sum_text += f"\nTotal sum of crimes in {selected_state_ut}:\n\n"
            yearly_data = total_crime_by_state[selected_state_ut]
            for year, total_sum in yearly_data.items():
                total_sum_text += f"Year {year}: {total_sum}\n"
        return total_sum_text

    # Function to plot bar graph for the selected state/UT
    def plot_graph():
        selected_state_ut = selected_state_ut_var
        yearly_data = total_crime_by_state[selected_state_ut]
        years = list(yearly_data.keys())
        total_sum = list(yearly_data.values())

        # Clear existing graph
        plt.clf()

        # Plot new graph
        plt.bar(years, total_sum, color='blue')
        plt.plot(years, total_sum, color='red', marker='o')
        plt.title(f"Total Crimes over the Years in {selected_state_ut}")
        plt.xlabel("Year")
        plt.ylabel("Total Crimes")
        plt.grid(True)

        # Display the plot
        st.pyplot()

    # Initialize an empty dictionary to store total sums for each state/UT
    total_crime_by_state = {}

    # List of all unique state/UT names (stripped of whitespace)
    states_ut_names = [state_ut.strip() for state_ut in crime['STATE/UT'].unique()]

    # Iterate over each state/UT
    for state_ut in states_ut_names:
        # Initialize an empty dictionary to store total sums for each year
        total_crime_by_year = {}

        # Iterate over the years from 2001 to 2012
        for year in range(2001, 2013):
            # Filter data for the current state/UT and year
            state_ut_yearly = crime[(crime['STATE/UT'] == state_ut) & (crime['Year'] == year)]

            # Sum the specified columns for the filtered data
            total_crime_by_year[year] = state_ut_yearly[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                                                          'Assault on women with intent to outrage her modesty',
                                                          'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
                                                          'Importation of Girls']].sum().sum()

        # Store the total sum of crimes for the current state/UT in the main dictionary
        total_crime_by_state[state_ut] = total_crime_by_year

    # Dropdown menu to select the state/UT
    selected_state_ut_var = st.selectbox("Select State/UT", states_ut_names)

    # Button to update the displayed text
    update_button = st.button("Update")
    if update_button:
        total_sum_text = update_display()
        st.text(total_sum_text)

    # Button to plot the bar graph
    plot_button = st.button("Plot Graph")
    if plot_button:
        plot_graph()


elif selected_tab == "Crime Wise Proportion":
    st.title("Crime Wise Proportion for Each State")

    # Function to plot the proportion of crimes for a selected state
    def plot_crime_proportion(selected_state):
        state_data = crime[crime['STATE/UT'] == selected_state]
        total_crimes = state_data[['Rape', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty',
                                   'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
                                   'Importation of Girls']].sum().sum()
        percentages = (state_data[['Rape', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty',
                                   'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
                                   'Importation of Girls']].sum() / total_crimes) * 100

        plt.figure(figsize=(10, 6))
        wedges, texts, autotexts = plt.pie(percentages, labels=percentages.index,
                                            colors=sns.color_palette('pastel'), startangle=90, autopct='%.1f%%')
        plt.title(f"Proportion of Crimes in {selected_state}")
        for autotext in autotexts:
            autotext.set_fontsize(11)
        st.pyplot()

    # Dropdown menu to select the state
    selected_state = st.selectbox("Select State/UT", crime['STATE/UT'].unique())

    # Button to plot the proportion of crimes
    plot_button = st.button("Plot Proportion of Crimes")
    if plot_button:
        plot_crime_proportion(selected_state)

```
